```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,4*np.pi,1000)
y=np.sin(x)
fig, ax=plt.subplots()
ax.plot(x,y)
fig.show
```

```python
x=np.random.uniform(0,1,100)
y=np.random.uniform(0,1,100)
ax.scatter(x,y)
```

```python
x=np.arange(10)
y=np.random.uniform(1,10,10)
ax.bar(x,y)
```

```python
z=np.random.uniform(0,1,(8,8))
ax.imshow(z)
```
